```python
# DataFrame 연결(행방향)
# 같은 컬럼끼리 붙는다.

import numpy as np
import pandas as pd

df1 = pd.DataFrame({'a':['a0', 'a1', 'a2', 'a3'],
                    'b':[1, 2, 3, 4],
                    'c':['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])
display(df1)

df2 = pd.DataFrame({'b':[5, 6, 7, 8],
                    'c':['c0', 'c1', 'c2', 'c3'],
                    'd':['d0', 'd1', 'd2', 'd3'],
                    'e':['e0', 'e1', 'e2', 'e3']},
                    index=[2, 3, 4, 5])
display(df2)

# 기존의 index를 유지하면서 연결
# pd.concat([연결할 DataFrmae들], axis=n)
# result_df = pd.concat([df1, df2], axis=0)
# display(result_df)

# 기존의 index를 없애고 숫자 index를 다시 설정
result_df = pd.concat([df1, df2], axis=0,
                      ignore_index=True)   # DataFrame을 연결하고 인덱스 다시 설정
display(result_df)
```

```python
# DataFrame의 결합(merge)
# inner join

import numpy as np
import pandas as pd

data1 = {'학번':[1, 2, 3, 4],
         '이름':['홍길동', '김연아', '아이유', '신사임당'],
         '학년':[3, 2, 1 ,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학번':[1, 2, 4, 5],
         '학과':['철학과', '수학과', '컴퓨터', '국어국문'],
         '학점':[1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2)
display(df2)

result_df = pd.merge(df1, df2, on='학번', how='inner')   # inner join
# 학번이 같은 것만 가져온다.
display(result_df)
```

```python
# DataFrame의 결합(merge)
# outer join
# inner join에 포함되지 않는 데이터들도 포함.

import numpy as np
import pandas as pd

data1 = {'학번':[1, 2, 3, 4],
         '이름':['홍길동', '김연아', '아이유', '신사임당'],
         '학년':[3, 2, 1 ,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학번':[1, 2, 4, 5],
         '학과':['철학과', '수학과', '컴퓨터', '국어국문'],
         '학점':[1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2)
display(df2)

# result_df = pd.merge(df1, df2, on='학번', how='outer')   # outer join
# display(result_df)

result_df = pd.merge(df1, df2, on='학번', how='left')   # left outer join
# 왼쪽에 있는 DataFrame의 조건에 포함되지 않는 데이터만 포함한다.
display(result_df)

result_df = pd.merge(df1, df2, on='학번', how='right')   # right outer join
# 오른쪽에 있는 DataFrame의 조건에 포함되지 않는 데이터만 포함한다.
display(result_df)
```

```python
# inner join
# 컬럼 명이 다른 경우

import numpy as np
import pandas as pd

data1 = {'학번':[1, 2, 3, 4],
         '이름':['홍길동', '김연아', '아이유', '신사임당'],
         '학년':[3, 2, 1 ,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학생학번':[1, 2, 4, 5],
         '학과':['철학과', '수학과', '컴퓨터', '국어국문'],
         '학점':[1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2)
display(df2)

# 결합하려는 기준 column명이 서로 다른 경우.
result_df = pd.merge(df1, df2, left_on='학번', right_on='학생학번', how='inner')
display(result_df)
```

```python
# column과 index를 이용하여 결합

import numpy as np
import pandas as pd

data1 = {'학번':[1, 2, 3, 4],
         '이름':['홍길동', '김연아', '아이유', '신사임당'],
         '학년':[3, 2, 1 ,4]}

df1 = pd.DataFrame(data1)
display(df1)

data2 = {'학과':['철학과', '수학과', '컴퓨터', '국어국문'],
         '학점':[1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2, index=[1, 2, 4, 5])
display(df2)

# 결합하려는 기준 column과 다른 DataFrame의 행 index를 결합하려면
# 어떻게 해야 할까?
result_df = pd.merge(df1, df2,
                     left_on='학번',
                     right_index=True,
                     how='inner')
display(result_df)
```

```python
# index와 index를 이용해서 결합

import numpy as np
import pandas as pd

data1 = {'이름':['홍길동', '김연아', '아이유', '신사임당'],
         '학년':[3, 2, 1 ,4]}

df1 = pd.DataFrame(data1, index=[1, 2, 3, 4])
display(df1)

data2 = {'학과':['철학과', '수학과', '컴퓨터', '국어국문'],
         '학점':[1.2, 3.3, 2.7, 4.0]}

df2 = pd.DataFrame(data2, index=[1, 2, 4, 5])
display(df2)

# index와 index를 이용해서 결합하기
result_df = pd.merge(df1, df2,
                     left_index=True,
                     right_index=True,
                     how='inner')
display(result_df)
```

```python
#### DataFrame 행과 열에 함수를 mapping
#### 예제로 사용할 데이터 셋은 titanic을 이용해보자

import numpy as np
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
# display(titanic.head())

df = titanic.loc[:,['age', 'fare']]
display(df.head())

# 함수는 apply()를 사용한다.  axis를 명시해야 한다.

# 정규화할 때 배운 Min-Max Scaling 처리를 해보자
# x(scaled) = (x - x(min)) / (x(max) - x(min))

def min_max(s):
    return (s - s.min()) / (s.max() - s.min())

# DataFrame에 apply를 적용할 때 axis=0 하면 열 단위.. axis=1 하면 행 단위
# 기존에 알던 것과 반대
result_df = df.apply(min_max, axis=0)
display(result_df.head())   # DataFrame에 대한 정규화를 진행.

def add_two_number(a, b):
    return a + b

df['add'] = df.apply(lambda x: add_two_number(x['age'], x['fare']), axis=1)  # 행 단위
display(df.head())
```
